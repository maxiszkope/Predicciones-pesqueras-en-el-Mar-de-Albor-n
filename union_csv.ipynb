### Crear el DataFrame básico para el análisis.

Definimos una función para unir los datos extraidos mediante Web Scrapping.

```python
def concatenar_csv_en_carpeta(ruta_carpeta):
    # Obtener la lista de archivos CSV en la carpeta.
    archivos_csv = glob.glob(os.path.join(ruta_carpeta, '*.csv'))
    
    # Leer cada archivo CSV en un DataFrame y agregarlo a una lista.
    lista_dfs = []
    for archivo in archivos_csv:
        df = pd.read_csv(archivo)
        lista_dfs.append(df)
    
    # Concatenar todos los DataFrames en la lista.
    df_concatenado = pd.concat(lista_dfs, ignore_index=True)
    
    return df_concatenado


ruta_carpeta = r'C:\Users\pejer\Desktop\BOOTCAMP\Analisis_Mediterraneo\Datos\Pesqueros\Limpios'
df_concatenado = concatenar_csv_en_carpeta(ruta_carpeta)

```

Transformar coordenadas de Puertos APPA.

```python
def transformar_coordenadas(coordenada, es_longitud=False):
    # Eliminar comillas, comas y apóstrofes.
    coordenada = coordenada.replace('"', '').replace("'", '').replace(',', '').replace('.', '')

    # Eliminar la letra (N o O).
    coordenada = coordenada.replace('N', '').replace('O', '')

    # Cambiar el símbolo de grados por un punto.
    coordenada = coordenada.replace('º', '.').replace('°', '.')

    # Eliminar espacios en blanco y dividir en partes.
    partes = coordenada.strip().split()

    # Unir las partes antes y después del punto.
    resultado = ''.join(partes[:-1])

    # Tomar solo los primeros 4 dígitos después del punto.
    resultado += partes[-1][:4]

    # Si es longitud y se especifica, multiplicar por -1.
    if es_longitud:
        resultado = str(float(resultado) * -1)

    return resultado


df = pd.read_csv('ruta a cuota_pesquera.csv')

# Aplicar la función a las columnas de longitud y latitud.
df['Latitud'] = df['Latitud'].apply(transformar_coordenadas)
df['Longitud'] = df['Longitud'].apply(transformar_coordenadas, es_longitud=True)

df.to_csv('cuota_pesquera.csv', index = False)
```
