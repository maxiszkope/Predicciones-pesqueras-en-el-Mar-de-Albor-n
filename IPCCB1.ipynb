Con el modelo obtenido en la plataforma de Azure e implementada su inferencia para la exportación del modelo, procedemos al calculo de las variables relacionadas mediante una clave de API y una url, las cuales alojan mi modelo.


Predicción de la clorofila en las condiciones futuras.

```python
def allowSelfSignedHttps(allowed):
    if allowed and not os.environ.get('PYTHONHTTPSVERIFY', '') and getattr(ssl, '_create_unverified_context', None):
        ssl._create_default_https_context = ssl._create_unverified_context

allowSelfSignedHttps(True)


def predict_clorofila(pH_value, Media_Temperatura_value, api_key):
    url = 'xxx'
    data = {
        "input_data": {
            "columns": ["pH", "Media_Temperatura"],
            "data": [[pH_value, Media_Temperatura_value]]
        }
    }
    body = str.encode(json.dumps(data))
    headers = {
        'Content-Type': 'application/json',
        'Authorization': 'Bearer ' + api_key
    }
    try:
        req = urllib.request.Request(url, body, headers)
        with urllib.request.urlopen(req) as response:
            result = json.loads(response.read())
            return result
    except urllib.error.HTTPError as error:
        error_message = f"Error en la solicitud: {error.code}\n{error.read().decode('utf8', 'ignore')}"
        return {"error": error_message}

API_KEY = 'xxx'

# Actualización de los nuevos valores en el DataFrame.
tqdm.pandas(desc="Obteniendo predicciones")
df['Media_Clorofila'] = df.progress_apply(lambda row: predict_clorofila(row['pH'], row['Media_Temperatura'], API_KEY), axis=1)
```
